```python
def create_table(table_name):
    mycon = mysql.connector.connect(
        host="localhost",
        user="root",
        passwd="2616",
        auth_plugin="mysql_native_password",
        database="result"
    )
    mycursor = mycon.cursor()
    mycursor.execute(f"CREATE TABLE IF NOT EXISTS {table_name}(rowid INT AUTO_INCREMENT PRIMARY KEY, stu_id VARCHAR(20) UNIQUE, name VARCHAR(20), admn_no INT(7), dob DATE, totalmarks INT(4), grade VARCHAR(2))")
    mycon.commit()
    
    print("Tables under result: ")
    mycursor.execute("SHOW TABLES")
    for x in mycursor:
        print(x)

def add_new_record(table_name, records):
    mycon = mysql.connector.connect(
        host="localhost",
        user="root",
        passwd="2616",
        auth_plugin="mysql_native_password",
        database="result"
    )
    mycursor = mycon.cursor()
    sql_check_existence = f"SELECT COUNT(*) FROM {table_name} WHERE stu_id = %s"  # Fix the SQL query here
    sql_insert = f"INSERT INTO {table_name}(stu_id, name, admn_no, dob, totalmarks, grade) VALUES(%s, %s, %s, %s, %s, %s)"
    
    for record in records:
        mycursor.execute(sql_check_existence, (record[0],))
        result = mycursor.fetchone()[0]

        if result == 0:
            mycursor.execute(sql_insert, record)  
            print(f"Record {record[0]} inserted successfully.")
        else:
            print(f"Record {record[0]} already exists, skipping insertion.")
    
    mycon.commit()
    print("All records added successfully.")

def remove_duplicates(table_name):
    mycon = mysql.connector.connect(
        host="localhost",
        user="root",
        passwd="2616",
        auth_plugin="mysql_native_password",
        database="result"
    )
    mycursor = mycon.cursor()

    duplicate_query = f"""
    SELECT stu_id, name, admn_no, dob, totalmarks, grade, COUNT(*)
    FROM {table_name}
    GROUP BY stu_id, name, admn_no, dob, totalmarks, grade
    HAVING COUNT(*) > 1
    """

    mycursor.execute(duplicate_query)
    duplicate_records = mycursor.fetchall()

    print("Redundant records found:")
    for record in duplicate_records:
        print(record)

    delete_query = f"""
    DELETE FROM {table_name}
    WHERE (stu_id, name, admn_no, dob, totalmarks, grade) IN (
        SELECT stu_id, name, admn_no, dob, totalmarks, grade
        FROM (
            SELECT stu_id, name, admn_no, dob, totalmarks, grade, ROW_NUMBER() OVER (PARTITION BY stu_id, name, admn_no, dob, totalmarks, grade) AS rn
            FROM {table_name}
        ) AS tmp
        WHERE rn > 1
    )
    """

    mycursor.execute(delete_query)
    mycon.commit()
    mycon.close()

    print("Duplicate records removed successfully.")

```
